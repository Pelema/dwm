```python
import pandas as pd
```

```python
df = pd.read_csv("./barbershop.csv")
df.head()
#df.info()
```

```python
avg = df['cost'].mean()
print(avg)

med = df['cost'].median()
print(med)


std = df['cost'].std()
print(std)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

sns.set(style="whitegrid")
plt.figure(figsize=(10, 8))
ax = sns.boxplot(x='shop_type', y='cost', data=df, orient="v")
```

```python
filter_data = df.dropna(subset=['cost'])
plt.figure(figsize=(14, 8))
sns.distplot(filter_data['cost'], kde=False)
```

```python
type_counts = df['promotion_platform_1'].value_counts()
print(type_counts)
df2 = pd.DataFrame({'prtype': type_counts}, 
                  index = ['b', 'o', 'r', 'u'])
print(df2)
df2.plot.pie(y='prtype', figsize=(10, 10), autopct='%1.1f%%')
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(20, 10))
ax = sns.countplot(x='hairstyle', data=df)
```
